```python
import matplotlib.pyplot as plt
import numpy as np

# Datos para los gráficos iniciales
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
categorias = ['A', 'B', 'C', 'D']
valores = [10, 20, 15, 25]

# Crear una figura con 4 subplots (2x2)
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Gráfico de líneas
axs[0, 0].plot(x, y1, label='Seno')
axs[0, 0].plot(x, y2, label='Coseno')
axs[0, 0].set_title('Gráfico de líneas')
axs[0, 0].legend(loc='upper right')
axs[0, 0].grid(True)
axs[0, 0].set_xlabel('Eje X')
axs[0, 0].set_ylabel('Eje Y')

# Gráfico de barras
axs[0, 1].bar(categorias, valores)
axs[0, 1].set_title('Gráfico de barras')
axs[0, 1].set_xlabel('Categorías')
axs[0, 1].set_ylabel('Valores')
axs[0, 1].grid(True, axis='y')

# Datos para el gráfico de dispersión
x_scatter = np.random.rand(100)
y_scatter = np.random.rand(100)

# Gráfico de dispersión
axs[1, 0].scatter(x_scatter, y_scatter, color='r', alpha=0.5)
axs[1, 0].set_title('Gráfico de dispersión')
axs[1, 0].set_xlabel('Eje X')
axs[1, 0].set_ylabel('Eje Y')
axs[1, 0].grid(True)

# Datos para el gráfico de caja
datos_boxplot = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Gráfico de caja
axs[1, 1].boxplot(datos_boxplot, vert=False)
axs[1, 1].set_title('Gráfico de caja')
axs[1, 1].set_xlabel('Valores')
axs[1, 1].set_yticklabels(['Desviación 1', 'Desviación 2', 'Desviación 3'])
axs[1, 1].grid(True)

# Mostrar todos los gráficos
plt.show()
```
